# Normal distribution
__Human's length example__

```python
# find a value in  a Normal Distribution with givens means and standard deviation
my_data = np.random.normal(loc = 170, scale = 5, size= 100000)
```

```python
#take multiple values
my_data
```

```python
#find minimum
my_data.min()
```

```python
#find maximum
my_data.max()
```

```python
#visualize
pd.DataFrame(my_data).hist()
```

```python
np.array([1,2,3])-3
```

## standard normal distribution

```python
#use the same generated normally distributed data
```

```python
#find mean
my_data.mean()
```

```python
#find standard deviation
my_data.std()
```

```python
#standardize the data
standar_data = [(i-my_data.mean())/ my_data.std() for i in my_data]
print(standar_data)
```

```python
# example difference between subtracting a constant from list and numpy|
#np.array([3,4,5])-3
```

```python
#visualize data

pd.DataFrame(standar_data).hist()
```

```python
#write function for z score
def z_score(sample,mean, std):
    return (sample-mean)/ std
```

```python
#calculate z score
player1 = z_score(12,9,2)
player2 = z_score(19,15,3)
player1,player2
```

__Example 2 :__

     A national achievement test is administered annually to 3rd graders. The test has a mean score of 100 and a standard deviation of 15. If Jane's z-score is 1.20, what was her score on the test?


```python
# write function to get original score from Z-score
def score_from_z_score(z_score,mean,std):
    return (z_score*std)+mean
```

```python
#calculate te original data
original_data = score_from_z_score(1.2,100,15)
```
