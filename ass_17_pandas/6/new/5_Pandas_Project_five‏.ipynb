```python
import pandas as pd
```

```python
df = pd.read_csv('3.1-data-sheet-udemy-courses-web-development.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df[df.duplicated()]
```

## there is no data duplicated and no data empity (2 cells that is not big deal)

```python
df['Rating'].unique()
```

### best cources based on rating and number of subscribers

```python
df[(df['Rating'] == df['Rating'].max())].sort_values(by='num_subscribers' ,ascending = False).head()
```

### longest course duration

```python
df[df['content_duration'] == df['content_duration'].max()]
```

### longest course number lectures

```python
df[df['num_lectures'] == df['num_lectures'].max()]
```

#### number of review in courses in all levels( that mean beginner level is most selling)

```python
df.groupby('level')['num_reviews'].max().to_frame().sort_values('num_reviews',ascending = False)
```

## highest price in all levels (that mean the level not depend on price)

```python
df.groupby('level')['price'].max().to_frame().sort_values('price',ascending = False)
```

```python
df.head()
```

```python
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
df.info()
```

### make column year

```python
df['year'] = df['published_timestamp'].dt.year
```

```python
df
```

### highest selling in 2013

```python
df.groupby('year')['num_subscribers'].max().to_frame().sort_values('num_subscribers',ascending = False)
```

### highest number of courses generated in 2016

```python
df.groupby('year')['course_title'].count().to_frame().sort_values('course_title',ascending = False)
```
