write a function to detect the outlier in the dataframe and replace with chosen method(mean/ median/drop/ nothing)

```python
import pandas as pd
import numpy as np
from datasist.structdata import detect_outliers
df = pd.read_csv('boston_house_prices.csv')
```

```python

def detect_deal_with_outlier(df, dealing_method='none'):
    
    """"
    the function takes the dataframe and loop on its columns and detect the outlier indices in each of them
    
    Input:
    df: type:dataFrame
    dealing_method:type:str: 'none', 'mean','median','drop'
        'none': the outliers indices will remain the same
        'mean': the outliers indices in each column will be replace by the mean value of its corresponding column 
        'median': the outliers indices in each column will be replace by the median value of its corresponding column
        'drop': the outliers indices will be dropped
    
    Output:
    outliers_indices: list of lists of outliers indices in each column in the dataframe
    filtered_dataframe: type:DataFrame     
    """
    E = dealing_method
    print(E)
    filtered_dataframe = df
    if E == 'none':
        outliers_indices = 0
        filtered_dataframe = df
    if E == 'mean':
        
        for i in filtered_dataframe.columns:
            outliers_indices = detect_outliers(filtered_dataframe,0,[i])
#             mean_col = df[i].mean()
            filtered_dataframe[i].iloc[outliers_indices] = filtered_dataframe[i].mean()
    if E == 'median':
        
        for i in filtered_dataframe.columns:
            outliers_indices = detect_outliers(filtered_dataframe,0,[i])
#             mean_col = df[i].mean()
            filtered_dataframe[i].iloc[outliers_indices] = filtered_dataframe[i].median()
    if E == 'drop':
        
        for i in filtered_dataframe.columns:
            outliers_indices = detect_outliers(filtered_dataframe,0,[i])
#             mean_col = df[i].mean()
            filtered_dataframe.drop(outliers_indices,inplace = True)
        
    return outliers_indices, filtered_dataframe
```
