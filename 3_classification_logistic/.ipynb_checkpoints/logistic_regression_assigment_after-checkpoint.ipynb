```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import plotly.express as px
```

```python
df = pd.read_csv('Breast Cancer Wisconsin (Diagnostic) Data Set .csv')
```

```python
df.head()
```

```python
df.drop(columns= ['Unnamed: 32'], inplace= True)
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
df.drop(columns= ['id'], inplace = True)
```

```python
px.imshow(df.corr(), text_auto= True)
```

```python
df.columns
```

```python
df.drop(columns= ['radius_mean','perimeter_mean'],inplace = True)
```

```python
X= df.drop(columns=['diagnosis'])
y= df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X ,y, random_state= 42, test_size= 0.3)
```

```python
X_train.head()
```

```python
scaler_train = MinMaxScaler()
scaler_train.fit(X_train)
scaler_train.data_max_
```

```python
X_train_trans= scaler_train.transform(X_train)
```

```python
scaler_test = MinMaxScaler()
scaler_test.fit(X_test)
scaler_test.data_max_
```

```python
X_test_trans= scaler_test.transform(X_test)
```

```python
model_with_scale= LogisticRegression()
```

```python
model_with_scale.fit(X_train_trans,y_train)
```

```python
y_predict= model_with_scale.predict(X_test_trans)
```

```python
y_predict
```

```python
# nice accuracy with scale
accuracy_score(y_test,y_predict)
```

```python
# without scale
model_2_without_scale = LogisticRegression()
model_2_without_scale.fit(X_train,y_train)
y_predict_without_scale= model_2_without_scale.predict(X_test)
accuracy_score(y_test,y_predict_without_scale)
```
