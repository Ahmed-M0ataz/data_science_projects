```python
#scrape every page in website
#scrape every next page in category in website
count=2
import time
import csv
import xlsxwriter
workbook = xlsxwriter.Workbook('books_exel.xlsx')
worksheet = workbook.add_worksheet('fristsheet')
data = ['name', 'price', 'rate']
worksheet.write_row('A1', data)

with open ('books.csv', 'w', newline='') as csv_file:
    writer= csv.DictWriter(csv_file, fieldnames= ['name','price','rate'])
    writer.writeheader()
    response = requests.get('https://books.toscrape.com/catalogue/category/books_1/index.html')
    soup = BeautifulSoup(response.content, 'html.parser')
    catalog = soup.find('ul',{'class': 'nav-list'}).find('ul').find_all('a')
    for page in catalog:
        response2 = requests.get('https://books.toscrape.com/catalogue/category'+page.get('href')[2:])

        soup = BeautifulSoup(response2.content, 'html.parser')

        writer.writerow({'name': '**********','price': page.text.split()[0],'rate': '**********'})
        worksheet.write_row('A'+str(count),['**********', page.text.split()[0], '**********'])
        count+= 1
        books = soup.find_all('article', {'class':'product_pod'})


        next_page = soup.find('ul',{'class': 'pager'})
        
        for book in books:
            name = book.find('h3').find('a').get('title')
            price = book.find('p',{"class": 'price_color'}).text[1:]
            rate = book.find('p',{"class": 'star-rating'}).get('class')
            if rate[1] == 'One':
                rate = 1
            elif rate[1] == 'Two':
                rate = 2
            elif rate[1] == 'Three':
                rate = 3
            elif rate[1] == 'Four':
                rate = 4
            elif rate[1] == 'Five':
                rate = 5
            else:
                rate = 'no rating'
            writer .writerow({'name': name,'price': price,'rate': rate})
            worksheet.write_row('A'+str(count),[name, price, rate])
            count+= 1
        if next_page is None:
            continue
        else:
            time.sleep(2.4)
            num_page = int(next_page.find('li',{'class':'current'}).text.split()[-1])
            print(num_page)
            for current_page in range(num_page-1):
                next_page_url = soup.find('ul',{'class': 'pager'}).find('li',{'class':'next'}).find('a').get('href')
                print(next_page_url)
                response_pages = requests.get('https://books.toscrape.com/catalogue/category'+page.get('href')[2:-10]+next_page_url)
                print('https://books.toscrape.com/catalogue/category'+page.get('href')[2:-10]+next_page_url)
                #print('https://books.toscrape.com/catalogue/category'+page.get('href')[2:])
                
                print(response_pages)
                
                soup = BeautifulSoup(response_pages.content, 'html.parser')
                #print(soup)
                books = soup.find_all('article', {'class':'product_pod'})
                for book in books:
                    name = book.find('h3').find('a').get('title')
                    price = book.find('p',{"class": 'price_color'}).text[1:]
                    rate = book.find('p',{"class": 'star-rating'}).get('class')
                    if rate[1] == 'One':
                        rate = 1
                    elif rate[1] == 'Two':
                        rate = 2
                    elif rate[1] == 'Three':
                        rate = 3
                    elif rate[1] == 'Four':
                        rate = 4
                    elif rate[1] == 'Five':
                        rate = 5
                    else:
                        rate = 'no rating'
                    writer.writerow({'name': name,'price': price,'rate': rate})
                    worksheet.write_row('A'+str(count),[name, price, rate])
                    count+= 1
                time.sleep(1.4)
                
                
                
            
print('done')    
```
